```python
import spacy # is a object oriented library
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("Dr. Strange loves pav bhaji of mumbai. Hulk loves chaat of delhi")
for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:  # ex doc.sents --> object and the functions etc
    for word in sentence:
        print(word)
```
